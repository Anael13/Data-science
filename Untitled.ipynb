```python
import tensorflow
import keras
import numpy as np
```

```python
x = np.array([100.0,2000.0,30000.0,400000.0,5000000.0,60000000.0,700000000.0,8000000000.0],dtype=float)
y = np.array([3.0,4.0,5.0,6.0,7.0,8.0,9.0,10.0],dtype=float)
```

```python
model = keras.Sequential([keras.layers.Dense(units=20,input_shape=[1])])
```

```python
model.compile(optimizer="sgd",loss="mean_squared_error")
```

```python
model.fit(x,y,epochs=10000)
```

```python
model.predict([18])
```
